```python
import matplotlib.pyplot as plt
import numpy as np
labels = ['anger', "disgust","fear","happy","sad","surprised","normal"]
train = [3995,436,4097,7215,4830,3171,4965]
valid = [958,111,1025,1774,1247,831,1233]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x-width/2, train, width, label="Train")
rects2 = ax.bar(x+width/2, valid, width, label="Valid")

ax.set_ylabel("The number of picture")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.grid()
```
